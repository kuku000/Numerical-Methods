```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-poster")
%matplotlib inline
```

```python
# step size
h = 0.1
# define grid
x = np.arange(-2, 4, h)
 
# compute function
y=(1/4)*(x**4) + (1/3)*(x**3) + 5*x

#建立一個function list
function=[]
#將代入x值的結果存入list
for i in range(np.size(y)):
    function.append(y[i])

# compute corresponding grid
x_diff = x[1:-1:]
 
# compute exact solution
exact_solutionexact_solution = 3*(x_diff**2) + 2*x_diff

#function of forward_diff
def my_forward_diff(function,h):
    #建立建立一個function_2 list
    function_2=[]
    #將forward_diff的結果存入list
    for i in range(np.size(function)-1):
         function_2.append((function[i+1]-function[i])/h)
    
    return function_2
        
 
    
    
     
#function of backward _diff    
def my_backward_diff(function_2,h):
    #建立建立一個function_3 list
    function_3=[]
    #將backward_diff的結果存入list
    for i in range(1,np.size(function_2)):
         function_3.append((function_2[i]-function_2[i-1])/h)
    
    return function_3
#function of sec_order_diff     
def sec_order_diff(function,h):
    #建立建立一個 function_sec_order list
    function_sec_order=[]
     #將sec_order_diff的結果存入list
    for i in range(1,np.size(function)-1):
        function_sec_order.append((function[i+1]-2*function[i]+function[i-1])/h**2)
    return function_sec_order 
        
        
 
```

```python
#呼叫my_forward_diff進行第一次difference
function_finite_difference_1=my_forward_diff(function,h) 
#呼叫my_backward_diff進行第二次difference
function_finite_difference_2=my_backward_diff(function_finite_difference_1,h)
#呼叫function_sec_order進行difference
function_sec_order=sec_order_diff(function,h)
 
```

```python
#製造一個12*8的圖    
plt.figure(figsize = (12, 8))
#畫出exact_solution的圖
plt.plot(x_diff,exact_solution,linewidth="12",\
label = "exact solution  ",color='darkblue')
#畫出 finite difference approximation的圖
plt.plot(x_diff,function_finite_difference_2,linewidth="5",\
label = "finite difference approximation",linestyle ="--",color='orange')
#畫出 second-order difference approximation的圖
plt.plot(x_diff,function_sec_order ,linewidth="13",\
label = " second-order approximation" ,marker="*",linestyle=" ",color='lime')
 
#用於指定圖例的位置(預設左上角)
plt.legend()
#把圖show出來
plt.show()
#計算finite_difference最大誤差
max_error_finite_difference = max(abs(exact_solution -  function_finite_difference_2 ))
#計算second_order difference最大誤差
max_error_second_order = max(abs(exact_solution -function_sec_order ))
print('error of finite difference:',max_error_finite_difference)
print('error of second-order:',max_error_second_order)
 
```
