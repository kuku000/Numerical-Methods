```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-poster")
%matplotlib inline

#得到 cos(x)  
def Cos(x): 
    return np.cos(x)
# 定義 step size
h=0.1
#計算間格及定義每0.1為一格
x = np.arange(0, 2*np.pi,h)

#實現 central differences的函式
def central_diff(x,h):
     return  ( Cos(x+h)- Cos(x-h))/(2*h)
#把計算出來central differences的陣列給Central_diff(方便)    
Central_diff=central_diff(x,h)
#x這個定義區間格數的陣列給x_diff(X軸) 
x_diff = x
#定義 exact_solution
exact_solution=-np.sin(x_diff)    
    

#製造一個12*8的圖    
plt.figure(figsize = (12, 8))
#畫出central differences approximation的圖
plt.plot(x_diff,Central_diff ,linewidth="8",\
label = "central difference approximation")
#畫出exact_solution的圖
plt.plot(x_diff,exact_solution,linewidth="4",\
label = "exact solution  ")
#用於指定圖例的位置(預設左上角)
plt.legend()
#把圖show出來
plt.show()
#計算最大誤差
max_error = max(abs(exact_solution - Central_diff))
#印出最大誤差
print("The maximum error is",max_error)
 
```
