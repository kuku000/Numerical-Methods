```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

```python
def curve(x):
    f = 4*(x**2) + (1/3)*x + 5
    return f
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
def trapezoid_rule(g):

 
    a =-2
    b = 4
    n = g
    h = (b - a) / (n - 1)
    x = np.linspace(a, b, n) 
    f = curve(x)
    #為此算法原因為中間的funtion(x)都會重複算到所以先把他們全部加起來乘以二
    I_trap = (h/2)*(f[0] +2 * sum(f[1:n-1]) + f[n-1])
    err_trap =128-I_trap
    
    print(f"I_trap:{math.floor(I_trap*1000)/1000}")
    print(f"err_trap:{-math.floor(-err_trap*1000)/1000}")

    x_curve = np.linspace(a, b, 1000)
    plt.plot(x_curve , curve(x_curve), linewidth=2, color= 'r')
    for i in range(len(x)-1):
        
        if i%3 ==0 :
            color_order= 'darkorange'
        elif i%2 ==1 :
            color_order ='forestgreen'
        else:
            color_order = 'royalblue'
        
        plt.fill_between([x[i],x[i+1]],[curve(x[i]),curve(x[i+1])],color=color_order,alpha = 0.5 ) 
        plt.xlim(-2,4)
        plt.ylim(0,70)
        

       

 




 

    
```

```python
 trapezoid_rule(3)
 
```

```python
 trapezoid_rule(11)
```
